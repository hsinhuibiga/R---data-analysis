```python
from scipy.stats import linregress
slope, intercept, r_value, p_value, std_err = linregress(df_diesel['DATE'],df_diesel['price'])
```

```python
from matplotlib import pyplot as plt 
%matplotlib inline
plt.title("linear regression of Diesel price")
plt.ylabel("price") 
plt.xlabel("DATE")
line = [slope*xi + intercept for xi in df_diesel['DATE']]
plt.plot(df_diesel['DATE'],line,'r-', linewidth=3)
plt.scatter(df_diesel['DATE'], df_diesel['price'])
plt.show()
```
